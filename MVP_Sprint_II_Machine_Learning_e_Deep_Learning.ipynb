```python
# configuração para não exibir os warnings
import warnings
warnings.filterwarnings("ignore")

# Imports necessários
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

**Imprimindo Dicionário dos Dados**

```python
info = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.names'
response = requests.get(info)
print(response.text.strip())
```

**Carga do dataset**

como o arquivo de dados não contem os nomes das colunas, é necessário informar os nomes, conforme o dicionário.

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
df = pd.read_csv(url, header=None, delimiter=r"\s+", names=column_names)

# Remover as amostras com valores ausentes na coluna "LSTAT"
df.dropna(subset=df.columns, inplace=True)

# Verifique se os valores ausentes foram removidos
print(df.isnull().sum())
```

**reservamos os últimos registros do dataset como "dados não vistos", que serão avaliados após o treinamento final.**

```python
dataset = df[:300]
dados_nao_vistos = df[300:]
```

**Separação em conjunto de treino e conjunto de teste com holdout**

```python
test_size = 0.20 # tamanho do conjunto de teste
seed = 55 # semente aleatória
np.random.seed(seed) # definindo uma semente global

# Separação em conjuntos de treino e teste
array = dataset.values
X = array[:,0:13]
y = array[:,13]

X_train, X_test, y_train, y_test = train_test_split(X, y,
    test_size=test_size, shuffle=True, random_state=seed) # holdout com estratificação

# Parâmetros e partições da validação cruzada
scoring = 'neg_mean_squared_error'
num_particoes = 10
kfold = KFold(n_splits=num_particoes, shuffle=True, random_state=seed)
```

# Funções que serão reutilizadas

```python
# processa sequência de estimadores (modelos) de machine learning ou pipelines em uma lista de tuplas contendo o 'nome' e o 'modelo/pipeline' correspondente.
# realiza a validação cruzada para cada modelo/pipeline na sequência, imprime os resultados e retorna os nomes dos modelos e os resultados das validações cruzadas.
def run_sequences(sequence):
    results = []
    names = []
    for name, model in sequence:
        cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
        results.append(cv_results)
        names.append(name)
        msg = "%s: %.3f (%.3f)" % (name, cv_results.mean(), cv_results.std()) # formatando para 3 casas decimais
        print(msg)

    return names, results

# cria um boxplot dos resultados da validação cruzada de diferentes modelos/pipelines
def box_plot_results(names, results, title):
    fig = plt.figure(figsize=(25,6))
    fig.suptitle(title)
    ax = fig.add_subplot(111)
    plt.boxplot(results)
    ax.set_xticklabels(names, rotation=90)
    plt.show()
```

```python
#calcula a média dos resultados para cada preditor e retorna o nome do preditor com o menor valor médio.
def find_best_predictor(names, results):
    # Calcula a média dos resultados para cada preditor
    mean_results = [np.mean(r) for r in results]

    # Encontra o índice do preditor com o valor médio mais próximo de zero
    best_index = np.argmin(np.abs(mean_results))

    # Obtém o nome e o valor médio do preditor com melhor desempenho
    best_predictor = names[best_index]
    best_mean = mean_results[best_index]

    # Imprime o resultado
    print("Melhor preditor:", best_predictor)
    print("Valor médio do erro quadrático médio negativo:", best_mean)

    return best_predictor
```

```python
#recebe dicionários de transformers e de estimators e gera todas as combinações de tuplas de 'pipeline_names' e 'pipelines'
def create_pipelines(transformers, estimators):
    pipelines = []

    for transform_name, transformer in transformers.items():
        for name, estimator in estimators.items():
            pipeline_name = f"{name}-{transform_name}"

            if transformer is None:
                pipeline = Pipeline([(name, estimator)])
            else:
                pipeline = Pipeline([(transform_name, transformer), (name, estimator)])

            pipelines.append((pipeline_name, pipeline))

    return pipelines
```

```python
# recebe valores de X, determina o escalador apropriado ao melhor cenário, e então retona o rescaledX
def transform_values(X, best_scenario, transformers):
    scaler_key = best_scenario.split('-')[1]
    scaler = transformers.get(scaler_key)

    if scaler is not None:
        scaler.fit(X)
        rescaledX = scaler.transform(X)
    else:
        rescaledX = X

    return rescaledX

# controi um modelo com base no estimador definido no melhor cenário, usando os melhores parametros descobertos através do uso de Grid.
def build_model(best_scenario, best_params):
    estimator_key = next(iter(best_params)).split("__")[0]
    estimator = estimators.get(estimator_key)

    # Remover prefixo 'XX__' dos nomes dos parâmetros
    stripped_best_params = {param.replace(estimator_key + '__', ''): value for param, value in best_params.items()}
    print(estimator_key)
    print(estimator)
    print(stripped_best_params)
    if estimator is not None:
        model = estimator.set_params(**stripped_best_params)
        return model
    else:
        raise ValueError("Estimador não encontrado!")

# imprime o MSE e retorna as predicoes com base em modelo construido com base no melhor cenário e melhores parametros, e dados reescalonados se for o caso.
def run_predictions(X, y, best_scenario, best_params, transformers):
    rescaledX = transform_values(X, best_scenario, transformers)
    model = build_model(best_scenario, best_params)
    model.fit(rescaledX, y)
    predictions = model.predict(rescaledX)
    mse = mean_squared_error(y, predictions)
    mse_neg = -mse
    print("Erro quadrático médio negativo (neg_mean_squared_error) para o dataset:", mse_neg)
    return predictions

# plota um comparativo do dados originais e das predicoes
def plot_predictions(y_nao_vistos, predictions, label):
    # Criar um array de índices para os dados não vistos
    indices = range(len(y_nao_vistos))

    # Plotar o gráfico de comparação
    plt.plot(indices, y_nao_vistos, label=label)
    plt.plot(indices, predictions, label='predictions')
    plt.xlabel('Índice')
    plt.ylabel('Valor')
    plt.legend()
    plt.title('Comparação entre ' + label + ' e predictions')
    plt.show()
```

#Seleção e preparação de estimadores (modelos) e transformers

```python
# Definindo os parâmetros do regressor base para o BaggingRegressor
bagging_regressor_base = DecisionTreeRegressor()
num_trees = 100
max_features = 3

# Criando os modelos para o VotingRegressor
voting_regressor_bases = []
voting_regressor_bases.append(('LR', LinearRegression()))
voting_regressor_bases.append(('cart', DecisionTreeRegressor()))
voting_regressor_bases.append(('svm', SVR()))


# Transformações que serão utilizadas
transformers = {
    'orig': None,
    'padr': StandardScaler(),
    'norm': MinMaxScaler()
}

# Algoritmos que serão utilizados
estimators = {
    'LR': LinearRegression(),
    'KNN': KNeighborsRegressor(),
    'CART': DecisionTreeRegressor(),
    'SVR': SVR(),
    'Bagging': BaggingRegressor(base_estimator=bagging_regressor_base, n_estimators=num_trees),
    'RF': RandomForestRegressor(n_estimators=num_trees),
    'ET': ExtraTreesRegressor(n_estimators=num_trees),
    'Ada': AdaBoostRegressor(n_estimators=num_trees),
    'GB': GradientBoostingRegressor(n_estimators=num_trees),
    'Vot': VotingRegressor(estimators=voting_regressor_bases)
}
```

#Comparação de Modelos

```python
names, results = run_sequences(estimators.items())
best_predictor = find_best_predictor(names, results)
box_plot_results(names, results, 'Comparação dos Modelos')
```

#Comparação dos Modelos - Dataset orginal, padronizado e normalizado

Agora serão criados pipelines para processar o dataset antes de submetê-lo a avaliação de cross-validation, utilizando transformers de escala padrão e normalizada.



```python
pipelines = create_pipelines(transformers, estimators)
names, results = run_sequences(pipelines)
best_predictor = find_best_predictor(names, results)
box_plot_results(names, results, 'Comparação dos Modelos - Dataset orginal, padronizado e normalizado')
```

#Ajuste de hiperparâmetros utilizando a função GridSearchCV para encontrar a melhor combinação de hiperparâmetros para os modelos SVR (Support Vector Regression).

```python
pipelines = []

# Definindo os componentes do pipeline
etr = ('ET', ExtraTreesRegressor())
standard_scaler = ('StandardScaler', StandardScaler())
min_max_scaler = ('MinMaxScaler', MinMaxScaler())

pipelines.append(('etr-orig', Pipeline(steps=[etr])))
pipelines.append(('etr-padr', Pipeline(steps=[standard_scaler, etr])))
pipelines.append(('etr-norm', Pipeline(steps=[min_max_scaler, etr])))

param_grid = {
    'ET__n_estimators': [25, 50, 100, 200],
    'ET__max_depth': [None, 10, 20],
}

best_scenario = None
best_params = None
best_score = None

# Prepara e executa o GridSearchCV
for name, model in pipelines:
    grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring=scoring, cv=kfold)
    grid.fit(X_train, y_train)
    # imprime a melhor configuração
    print("Cenário: %s - Melhor: %f usando %s" % (name, grid.best_score_, grid.best_params_))
    if best_score is None or grid.best_score_ > best_score:
        best_scenario = name
        best_params = grid.best_params_
        best_score = grid.best_score_

print("Melhor cenário:", best_scenario)
print("Melhores parâmetros:", best_params)
```

```python
# Avaliação do modelo com o conjunto de testes
predictions = run_predictions(X_train, y_train, best_scenario, best_params, transformers)
plot_predictions(y_train, predictions, 'y_treinamento')
```

```python
# Avaliação do modelo com o conjunto de dados não vistos pelo modelo.
X_nao_vistos = dados_nao_vistos.iloc[:, :-1].values
y_nao_vistos = dados_nao_vistos.iloc[:, -1].values

predictions = run_predictions(X_nao_vistos, y_nao_vistos, best_scenario, best_params, transformers)
plot_predictions(y_nao_vistos, predictions, 'y_nao_vistos')
```

#Aplicando técnica de feature selection

Podemos realizar a selecao de features visando reducao da dimensionalidade, o que facilita a execucao do modelo e pode reduzir problemas de over-fitting.

```python
from sklearn.feature_selection import SelectFromModel

# Criação do modelo
modelo = build_model(best_scenario, best_params)
#modelo = ExtraTreesRegressor(max_depth=None, n_estimators=50)

# Seleção de atributos com base na importância
selector = SelectFromModel(modelo, threshold=-np.inf, max_features=1)
selector.fit(X_train, y_train)

# Reduz X para os atributos selecionados
features = selector.transform(X_train)

# Print dos resultados
print("Atributos Originais:", dataset.columns[0:13])

# Exibe os atributos selecionados
print("\nAtributos Selecionados:", selector.get_support(indices=True))
print("\nNomes dos Atributos Selecionados:", dataset.columns[selector.get_support(indices=True)].tolist())
```

```python
# Avaliação do modelo com o conjunto de dados não vistos pelo modelo, usando feature selection.

X_nao_vistos_feature_selected = selector.transform(X_nao_vistos)
predictions = run_predictions(X_nao_vistos_feature_selected, y_nao_vistos, best_scenario, best_params, transformers)
plot_predictions(y_nao_vistos, predictions, 'y_nao_vistos')
```

# The Tripadvisor Reviews 2023 Dataset

#Estimação do Rating de avaliçoes do TripAdvisor Utilizando Técnicas e Bibliotecas de Deep Learning

# **Objeto do trabalho**:

O objetivo deste trabalho é treinar um modelo de deep learning para predizer o rating (avaliação) postado pelos usuários do TripAdvisor com base nos reviews apresentados.


# **Detalhes sobre o dataset**:

O dataset utilizado é o "The TripAdvisor Reviews 2023 Dataset", que contém avaliações e reviews de usuários do TripAdvisor, em inlgês. O dataset é composto por um arquivo CSV contendo colunas "review_full" (o texto completo do review) e "rating_review" (o rating dado pelo usuário).

# **Implementação:**

**Estratégia geral**

Inicialmente, é testado o primeiro modelo com uma amostra balanceada pre-processada contendo cerca de 2.500 registros.    Posteriormente, esse modelo é aperfeiçado através da seleção de hiperparametros.

Em seguida, é criado um novo modelo usando os melhores parâmetros obtidos no passo anterior, porém agora treinado com um número maior de registros, já pre-processados (cerca de 25.000).   Esse modelo é utilizado para predizer os ratings do dataset completo.

Nos processos de treinamento da rede neural é utilizada uma condição de parada (EarlyStopping) em função do erro (loss).

**Seleção de amostra balanceada**

Como o dataset possui quase 150 mil avaliações, se tornaria muito demorado a utilizaçao do dataset completo para fins de treinamento de modelo de deep learning.  Por isso adotou-se como estratégia a seleção de uma amostra balanceada para treinamento do modelo. A amostra balanceada é construida por selecao aleatória de registros de modo que cada classe alvo (rating) tenha aproximadamente o mesmo número de registros.

**Pré-processamento**

Antes de se proceder ao treinamento, é realizado o pré-processamento do dataset de amostras balanceadas, o que incluiu:  *remoção de pontuações*, *tokenização*, *remoção de stopwords* e *aplicação de stemming nos reviews*.

**Hold-out e Tokenizacao**

O dataset processado de amostras balanceadas foi então dividido em conjunto de treinamento e teste, e as avaliações dos usuários foram convertidas em sequências de tokens para alimentar o modelo de deep learning.

**Modelo de Deep Learning**

O modelo utiliza uma camada de embedding, uma camada Conv1D, uma camada GlobalMaxPooling1D, uma camada de DropOut e uma camada densa para realizar a predição do rating.

>**Camada de Embedding**: A camada de embedding é responsável por mapear as palavras de entrada para vetores de números reais, também conhecidos como embeddings.  Cada palavra é representada por um vetor de tamanho output_dim, onde cada dimensão do vetor captura informações semânticas sobre a palavra.
Esses embeddings são aprendidos durante o treinamento do modelo e ajudam a capturar relações semânticas entre as palavras.

>>**input_dim**: Número máximo de palavras no vocabulário. Este parâmetro determina o tamanho do vetor de palavras de entrada.

>>**output_dim**: Dimensão do espaço vetorial no qual as palavras serão incorporadas. Cada palavra será representada por um vetor de tamanho output_dim.

>>**input_length**: Comprimento máximo da sequência de entrada (número de palavras em uma sequência).
Relação com os inputs da função:

>>**max_words**: Número máximo de palavras no vocabulário.
embedding_dim: Dimensão do espaço vetorial para a incorporação das palavras.

>>**max_sequence_length**: Comprimento máximo da sequência de entrada.

>**Camada de Conv1D:**  A camada de convolução 1D aplica filtros convolucionais à sequência de embeddings de palavras.  Os filtros convolucionais são movidos ao longo da sequência de embeddings para extrair recursos locais, capturando padrões e informações relevantes.  A ativação 'relu' é aplicada à saída da convolução para introduzir não-linearidade e permitir que o modelo aprenda representações mais complexas.

>>**filters**: Número de filtros convolucionais a serem aplicados.

>>**kernel_size**: Tamanho do kernel convolucional.

>>**activation**: Função de ativação aplicada à saída da camada.

> **Camada de GlobalMaxPooling1D:** A camada de GlobalMaxPooling1D realiza uma operação de pooling sobre as saídas da camada de convolução.  Ela reduz a dimensionalidade dos recursos extraídos, mantendo as características mais importantes.  O pooling máximo retorna o valor máximo de cada recurso ao longo da dimensão temporal, resultando em uma representação fixa e compacta dos recursos.

>**Camada de Dropout**:  A camada de Dropout é usada para evitar overfitting, que ocorre quando o modelo se ajusta demasiadamente aos dados de treinamento e tem dificuldade em generalizar para novos dados.  Durante o treinamento, os neurônios de entrada da camada de Dropout são desativados aleatoriamente com uma taxa de dropout_rate, o que ajuda a prevenir a dependência excessiva entre neurônios.  Essa técnica de regularização ajuda a melhorar a capacidade do modelo de generalizar para dados não vistos durante o treinamento.

>> **rate**: Taxa de dropout, que especifica a fração dos neurônios de entrada a serem zerados durante o treinamento.

>**Camada Dense**:  A camada Dense é uma camada de neurônios totalmente conectada.  Ela recebe as saídas do pooling e realiza uma transformação linear seguida de uma função de ativação.  Nesse caso, a função de ativação 'softmax' é usada para atribuir probabilidades às diferentes classes de saída.  A camada Dense final tem units=5, o que significa que ela produzirá uma saída com 5 dimensões, correspondendo às 5 classes possíveis do problema de classificação.

>>**units**: Número de neurônios na camada densa.

>>**activation**: Função de ativação aplicada à saída da camada.

O otimizador é definido como '*adam*', que é um otimizador popular baseado em gradiente descendente estocástico.

A função de perda é definida como '*sparse_categorical_crossentropy*', que é uma função de perda adequada para problemas de classificação com várias classes.
A métrica é definida como 'accuracy', que é usada para avaliar o desempenho do modelo durante o treinamento.

**Seleção dos melhores parametros**

Os melhores parâmetros do modelo foram encontrados utilizando uma busca em grid. Por fim, os resultados foram avaliados utilizando o erro quadrático médio (MSE) e uma plotagem comparando os ratings originais e preditos.

# 2 - Funcoes que irao ser utilizadas neste notebook

```python
def criar_dataset_amostras_balanceado(df, coluna_classe, tamanho_amostra_por_classe):
    # Obter as classes únicas presentes no dataframe
    classes_unicas = df[coluna_classe].unique()

    # Lista para armazenar as amostras balanceadas
    amostras_balanceadas = []

    # Para cada classe, selecionar uma amostra aleatória balanceada
    for classe in classes_unicas:
        # Obter as amostras correspondentes a essa classe
        amostras_classe = df[df[coluna_classe] == classe]

        # Verificar se o tamanho mínimo de amostras não excede o número total de amostras nessa classe
        tamanho_amostra = min(tamanho_amostra_por_classe, len(amostras_classe))

        # Selecionar uma amostra aleatória dessa classe com o tamanho mínimo
        amostras_aleatorias = amostras_classe.sample(tamanho_amostra, random_state=42)

        # Adicionar as amostras selecionadas à lista balanceada
        amostras_balanceadas.extend(amostras_aleatorias.values.tolist())

    # Converter a lista balanceada em um novo dataframe
    df_amostras_balanceado = pd.DataFrame(amostras_balanceadas, columns=df.columns)

    return df_amostras_balanceado
```

```python
def remove_pontuacao(texto):
    # Remoção de pontuações e sinais especiais, mantendo os sinais ! e ?
    texto_sem_pontuacao = re.sub(r'[^\w\s!?]', '', texto)
    return texto_sem_pontuacao

def tokenizacao(texto):
    # Tokenização do texto
    tokens = nltk.word_tokenize(texto.lower())
    return tokens

def remove_stopwords(tokens):
    # Remoção de stopwords (em inglês)
    stop_words = set(stopwords.words('english'))
    tokens_sem_stopwords = [token for token in tokens if token not in stop_words]
    return tokens_sem_stopwords

def stemming(tokens):
    # Stemming
    stemmer = PorterStemmer()
    tokens_stemmed = [stemmer.stem(token) for token in tokens]
    return tokens_stemmed

def preprocessar_texto(texto):
    texto_sem_pontuacao = remove_pontuacao(texto)
    tokens = tokenizacao(texto_sem_pontuacao)
    tokens_sem_stopwords = remove_stopwords(tokens)
    tokens_stemmed = stemming(tokens_sem_stopwords)
    texto_preprocessado = ' '.join(tokens_stemmed)
    return texto_preprocessado

def preprocess_dataframe(df):
    # Copia o dataframe original
    df_preprocessado = df.copy()

    # Aplica o pré-processamento em cada review
    reviews_preprocessadas = []
    for review in tqdm(df_preprocessado['review_full'], desc='Pré-processando'):
        reviews_preprocessadas.append(preprocessar_texto(review))

    df_preprocessado['review_full'] = reviews_preprocessadas

    return df_preprocessado
```

```python
def vocabulario(df, coluna):
    # Obter todas as avaliações em uma lista
    reviews = df[coluna].values.tolist()

    # Tokenização das palavras nas avaliações
    tokens = [word for review in reviews for word in review.split()]

    # Contagem das palavras únicas
    word_count = Counter(tokens)

    # Número de palavras únicas
    num_unique_words = len(word_count)

    print(f"O número de palavras únicas na coluna {coluna} é: {num_unique_words}")

    return num_unique_words
```

```python
# Função para criar o modelo da rede neural
def create_model(max_words, embedding_dim, dropout_rate, max_sequence_length):
    model = Sequential()
    model.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_sequence_length))
    model.add(Conv1D(128, 5, activation='relu'))
    model.add(GlobalMaxPooling1D())
    model.add(Dropout(dropout_rate))  # Adicionando a camada de Dropout
    model.add(Dense(units=5, activation='softmax'))
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
def criar_matriz_confusao(y_test, y_pred):
    # Criar a matriz de confusão
    cm = confusion_matrix(y_test, y_pred)

    # Verificar se há valores maiores que 999
    if np.max(cm) > 999:
        label_format = lambda x: f"{x/1000:.0f}K" if x > 999 else str(x)  # Utilizar unidades de mil (K)
        cm = cm / 1000  # Dividir todos os valores por 1000 para exibir em K
    else:
        label_format = str  # Utilizar formato decimal

    # Criar o heatmap da matriz de confusão
    sns.heatmap(cm, annot=True, fmt='', cmap="YlGnBu")

    # Definir os rótulos dos eixos
    classes = np.unique(y_test)
    plt.xticks(np.arange(len(classes)) + 0.5, classes)
    plt.yticks(np.arange(len(classes)) + 0.5, classes)

    # Adicionar rótulos dos valores
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            value = label_format(cm[i, j])
            color = "black" if cm[i, j] < np.max(cm) / 2 else "white"
            plt.text(j + 0.5, i + 0.5, value, ha="center", va="center", color=color)

    # Adicionar título e rótulos dos eixos
    plt.title("Matriz de Confusão")
    plt.xlabel("Classe Predita")
    plt.ylabel("Classe Real")

    # Exibir o gráfico
    plt.show()
```

# 3 - Elaboração do primeiro modelo de Deep Learning, usando uma amostra reduzida

Neste primeiro teste, iremos gerar um dataset amostras balanceadas (reduzido), em que cada classe de avaliacao (rating) deverá ter cerca de 500 amostras.

```python
tamanho_amostras_por_classe = 500

dataset_amostras_balanceado = criar_dataset_amostras_balanceado(df, 'rating_review', tamanho_amostras_por_classe)
print(dataset_amostras_balanceado)
```

**Teste do primeiro modelo**

A criação do modelo utiliza a métrica 'sparse_categorical_crossentropy', que implica as saídas devem ser valores discretos entre 0 e 4.   Dessa forma, deve ser feito um ajuste nas entradas usando o LabelEncoder, a fim de que as entradas sejam alteradas de 1 a 5 para 0 a 4.

Após treinamento, é necessário fazer a operação inversa do LabelEncoder para comparar as predições com os valores atuais.

```python
# Separando as avaliações e os ratings
reviews = dataset_amostras_balanceado_preprocessado['review_full'].values
ratings = dataset_amostras_balanceado_preprocessado['rating_review'].values

# Dividindo o dataset em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(reviews, ratings, test_size=0.2, random_state=42)

# Definindo parâmetros
vocab_size = vocabulario(dataset_amostras_balanceado_preprocessado, 'review_full')
embedding_dim = 50
dropout_rate = 0.2

# Tokenização das avaliações
tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(X_train)

# Convertendo as avaliações em sequências de tokens
X_train_sequences = tokenizer.texts_to_sequences(X_train)
X_test_sequences = tokenizer.texts_to_sequences(X_test)

# Determinando o tamanho máximo das sequências
max_sequence_length = max(len(sequence) for sequence in X_train_sequences)

# Padding das sequências para o mesmo tamanho
X_train_padded = pad_sequences(X_train_sequences, maxlen=max_sequence_length, padding='post')
X_test_padded = pad_sequences(X_test_sequences, maxlen=max_sequence_length, padding='post')

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode labels to integers
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

model = create_model(max_words=vocab_size, embedding_dim=embedding_dim, dropout_rate=dropout_rate, max_sequence_length=max_sequence_length)

# Imprimir resumo do modelo
print(model.summary())

# Definindo o callback EarlyStopping
early_stopping = EarlyStopping(monitor='loss', patience=8, min_delta=0.05)

history = model.fit(X_train_padded, y_train_encoded, epochs=20, batch_size=128, validation_split=0.2, callbacks=[early_stopping])

# Avaliação de Resultados
y_pred_encoded = model.predict(X_test_padded).argmax(axis=1)
y_pred = label_encoder.inverse_transform(y_pred_encoded)
mse = mean_squared_error(y_test, y_pred)
print("Erro quadrático médio (MSE):", mse)
```

**Avaliação do primeiro modelo**

```python
criar_matriz_confusao(y_test, y_pred)

# Calcule a acurácia
accuracy = accuracy_score(y_test, y_pred)

print("\nNesse primeiro teste, o modelo atingiu uma acurácia de ", accuracy)
```

Avaliacao de Hiper-parametros e seleção do melhor modelo (Best Model)

A biblioteca Keras Tuner serve para otimizar automaticamente os hiperparâmetros de modelos de aprendizado de máquina em Keras. Ela automatiza o processo de busca pelos melhores hiperparâmetros, melhorando o desempenho e a capacidade de generalização do modelo. Isso é feito através de algoritmos de busca, como o RandomSearch, que exploram diferentes combinações de hiperparâmetros para encontrar a melhor configuração. Essa biblioteca economiza tempo e esforço ao encontrar automaticamente os hiperparâmetros ideais para o seu modelo.

```python
# Usando uma versao modificada da funcao de criacao de modelo de deeep learning para receber os parametros vindos do turner
def create_model_for_keras_turner(hp):
    embedding_dim = hp.Choice('embedding_dim', values=[32, 64, 128])  # Escolha de valores possíveis para embedding_dim
    model = keras.Sequential()
    model.add(keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_sequence_length))
    model.add(keras.layers.Conv1D(128, 5, activation='relu'))
    model.add(keras.layers.GlobalMaxPooling1D())
    model.add(keras.layers.Dropout(rate=hp.Float('dropout_rate', min_value=0.2, max_value=0.4, step=0.1)))  # Adicionando a camada de Dropout
    model.add(keras.layers.Dense(units=5, activation='softmax'))
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

tuner = RandomSearch(
    create_model_for_keras_turner,
    objective='val_loss',
    max_trials=4,
    executions_per_trial=1,
    directory='my_dir',
    project_name='my_project'
)

tuner.search(X_train_padded, y_train_encoded, epochs=8, validation_split=0.2)

# Obter os melhores hiperparâmetros e criar o modelo final
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_model = tuner.hypermodel.build(best_hps)

history = best_model.fit(X_train_padded, y_train_encoded, epochs=20, batch_size=128, validation_split=0.2, callbacks=[early_stopping], verbose=1)
```

```python
# Melhor modelo encontrado
print("\nMelhores parâmetros:")
print(best_hps.values)
print("\nMelhor modelo:")
print(best_model.summary())
```

```python
# Avaliação de Resultados
y_pred_encoded = best_model.predict(X_test_padded).argmax(axis=1)
y_pred = label_encoder.inverse_transform(y_pred_encoded)
mse = mean_squared_error(y_test, y_pred)
print("\nErro quadrático médio (MSE):", mse)
```

```python
criar_matriz_confusao(y_test, y_pred)

# Calcule a acurácia
accuracy = accuracy_score(y_test, y_pred)

print("\nUsando o Best Model o modelo atingiu uma acurácia de ", accuracy)
```
